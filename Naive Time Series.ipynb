```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/daily_min_temp.csv', index_col='Date', parse_dates=['Date'])
df.head()
```

```python
df.Temp.plot()
```

```python
df = df.resample(rule='M').mean()
df
```

```python
df.Temp.plot(figsize=(15, 5))
```

## Naive Forecasting

### One Shift

```python
pred = df.copy()
pred = pred.shift()
```

```python
df['One Shift Pred'] = pred['Temp']
df
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df, "b-")
plt.plot(pred, 'r-')
```

### Mean Forecasting

```python
pred = df.copy()
pred.Temp = [df.loc[:date].Temp.mean() for date in df.index]
pred = pred.shift()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df['Temp'], "b-")
plt.plot(pred['Temp'], 'r-')
```

```python
df['Mean Forecasting'] = pred['Temp']
```

```python
df
```

### Moving Average

```python
pred = df.copy()
pred.Temp = df['Temp'].rolling(3).mean()
pred = pred.shift()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df['Temp'], "b-")
plt.plot(pred['Temp'], 'r-')
```

```python
df['Moving Average'] = pred['Temp']
```

```python
df
```

```python
plt.figure(figsize=(15, 5))

plt.plot(df['Temp'], "b-", label='Temp')
plt.plot(df['One Shift Pred'], 'r-', label='One Shift Pred')
plt.plot(df['Mean Forecasting'], 'g-', label='Mean Forecasting')
plt.plot(df['Moving Average'], 'y-', label='Moving Average')
plt.legend()
plt.show()
```
